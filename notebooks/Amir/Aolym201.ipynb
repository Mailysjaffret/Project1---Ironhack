```python
df1 = df.drop(columns=["edition_id"])
```

```python
df1 = df1[df1['year'] >= 1984]
```

```python
df1
```

```python
df1["edition"].unique()
```

```python
Paris = pd.read_csv('../../Project1---Ironhack/data/raw/medals_total.csv')
Paris
```

```python
Paris.rename(columns= {'Gold Medal':'gold', 'Silver Medal': 'silver', 'Bronze Medal': 'bronze', 
                       'Total': 'total', 'country_code': 'country_noc'}, inplace=True)
Paris
```

```python
Paris['year'] = 2024
Paris['edition'] = "2024 Summer Olympics"
Paris
```

```python
df2 = pd.concat([df1, Paris], axis=0).reset_index(drop=True)
df2
```

```python
# Create a mapping of country codes to country names
country_map = df2.dropna(subset=['country']).set_index('country_noc')['country'].to_dict()

# Fill missing values in "country" using the "country code" mapping
df2['country'] = df2['country'].fillna(df2['country_noc'].map(country_map))
```

```python
df2.loc[df2['country_noc'] == 'ALB', 'country'] = 'Albania'
df2.loc[df2['country_noc'] == 'AIN', 'country'] = 'Individual Neutral Athletes'
df2.loc[df2['country_noc'] == 'LCA', 'country'] = 'Saint Lucia'
df2.loc[df2['country_noc'] == 'DMA', 'country'] = 'Dominica'
df2.loc[df2['country_noc'] == 'CPV', 'country'] = 'Cape Verde'
df2.loc[df2['country_noc'] == 'EOR', 'country'] = 'Refugee Olympic Team'
```

```python
filtered_df = df2[df2['country_noc'] == 'EOR']
filtered_df
```

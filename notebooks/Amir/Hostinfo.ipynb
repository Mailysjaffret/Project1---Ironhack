```python
import pandas as pd
```

```python
host_info = pd.read_csv('data/raw/Olympics_Games.csv')
host_info
```

```python
host_info['year'].unique()
```

```python
host_info = host_info[(host_info['year'] >= 1984) & (host_info['year'] <= 2024)].reset_index(drop=True)

host_info
```

```python
host_info.drop(['edition_url', 'country_flag_url','edition_id', 'start_date', 'end_date', 'competition_date', 'isHeld'], axis=1, inplace=True)
```

```python
host_info
```

```python
medal_total = pd.read_csv('data/raw/Olympic_Games_Medal_Tally.csv')
medal_total
```

```python
medal_total = medal_total[medal_total['year'] >= 1984]
medal_total
```

```python
medal_total.drop(['edition_id'], axis=1, inplace=True)
medal_total
```

```python
medal_total['year'].unique()
```

```python
medal_total['country_noc'].unique()
```

```python
medal_total.info()
```

```python
total_medal = medal_total.groupby('country')['total'].sum().sort_values(ascending=False)
total_medal.head(12)
```

```python
grouped_data = medal_total.groupby(['year', 'country'])['total'].sum().reset_index()
grouped_data
```

```python
medal_total['country_noc'].unique()
```

```python
usa_df = medal_total[medal_total['country_noc'] == 'USA'].sort_values(by='year', ascending=True).reset_index()
usa_df
```

```python
grouped_by_year = usa_df.groupby('year').agg({'total': 'sum'}).reset_index()
grouped_by_year
```

```python
usa_medals_with_host_info = pd.merge(grouped_by_year, 
                                     host_info[['year', 'country_noc']], 
                                     on='year', 
                                     how='left')

usa_medals_with_host_info
```

```python
usa_medals_with_host_info['USA_host'] = usa_medals_with_host_info['country_noc'] == 'USA'

usa_medals_with_host_info.drop('country_noc', axis=1, inplace=True)

usa_medals_with_host_info['USA_host'].fillna(False, inplace=True)
```

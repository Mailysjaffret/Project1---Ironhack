```python
#Calculating the mean of each country total medals

medals_agg = medal_total.groupby('country_noc')['total'].agg(['mean', 'median']).sort_values(by=['mean', 'median'], ascending=False)

medals_agg['mean'] = medals_agg['mean'].astype(int)
medals_agg['median'] = medals_agg['median'].astype(int)
```

### Aggregates of all the country 

```python
medals_agg 
```

```python
#Picking up the countries that hosted more than once for further analysis of impact on medals because of hosting

countries = ['USA', 'KOR', 'CHN', 'JPN', 'FRA', 'CAN']
filtered_medals_agg = medals_agg.loc[countries]
filtered_medals_agg = filtered_medals_agg.sort_values(by='mean', ascending=False)
display(filtered_medals_agg) #The Aggregates of selected country for host and nonhost years
```

```python
countries = ['USA', 'KOR', 'CHN', 'JPN', 'FRA', 'CAN']

filtered_medal_total = medal_total[medal_total['country_noc'].isin(countries)].reset_index()
```

### Filtered countries total medals 

```python
filtered_medal_total
```

```python
#Adding a new column to mention hosting year
filtered_host_info = host_info[host_info['country_noc'].isin(countries)]
filtered_host_info = filtered_host_info.drop('city', axis=1).reset_index()
filtered_host_info['is_host'] = 'Host'

#filtered host_info DataFrame
filtered_host_info
```

```python
#Filtering data to see how many medals won as host and as non-host

medals_with_host_flag = pd.merge(filtered_medal_total, 
                                 filtered_host_info[['year', 'country_noc', 'is_host']], 
                                 on=['year', 'country_noc'], 
                                 how='left')

medals_with_host_flag['is_host'].fillna('Not Host', inplace=True)

medals_summary = medals_with_host_flag.groupby(['country_noc', 'is_host'])['total'].sum().reset_index()
```

### Count of medals as host vs non-host

```python
medals_summary
```

```python
#creating a pivot to display the values

medals_pivot = medals_summary.pivot_table(index='country_noc', columns='is_host', values='total', fill_value=0)
medals_pivot = medals_pivot.astype(int)
```

```python
#Modifying the pivot to add mean values of the total medals 

data = {'Host Avg': [41, 115, 38, 68, 50, 317],
        'Not Host Avg': [354, 598, 398, 302, 298, 1027]}
index = ['CAN', 'CHN', 'FRA', 'JPN', 'KOR', 'USA']
pivot_df = pd.DataFrame(data, index=index)

divisors = {
    'Host Avg': {'CAN': 2, 'CHN': 2, 'FRA': 2, 'JPN': 2, 'KOR': 2, 'USA': 3},
    'Not Host Avg': {'CAN': 7, 'CHN': 7, 'FRA': 7, 'JPN': 7, 'KOR': 7, 'USA': 6}
}

for col in pivot_df.columns:
    for country in pivot_df.index:
        divisor = divisors[col][country]
        pivot_df.at[country, col] = pivot_df.at[country, col] / divisor

pivot_df = pivot_df.astype(int).sort_values(by='Host Avg', ascending=False)
```

```python
import matplotlib.pyplot as plt

data = {
    'Host': [105, 57, 34, 25, 20, 19],
    'Not Host': [171, 85, 43, 42, 50, 56]
}
index = ['USA', 'CHN', 'JPN', 'KOR', 'CAN', 'FRA']
df = pd.DataFrame(data, index=index)

#Plotting
fig, ax = plt.subplots(figsize=(10, 7))

x = np.arange(len(df.index))  # the label locations
width = 0.35  # the width of the bars

#Updated color codes 
maroon_color = '#800000'  
new_golden_color = '#cca667'

#Creating the bars
bars1 = ax.bar(x - width/2, df['Host'], width, label='Host', color=maroon_color)
bars2 = ax.bar(x + width/2, df['Not Host'], width, label='Not Host', color=new_golden_color)

#Loop to add a label on top of each bar
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  
                    textcoords="offset points",
                    ha='center', va='bottom')

ax.set_xlabel('Country')
ax.set_ylabel('Average Medals Won')
ax.set_title('Average Medals Won by Country - Host vs Not Host')
ax.set_xticks(x)
ax.set_xticklabels(df.index)
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Data
countries = ['United States', 'China', 'Germany', 
             'Russian Fed', 'France', 'Great Britain', 
             'Italy', 'Canada', 'Australia', 'Japan']
total_medals = [1344, 713, 667, 543, 436, 421, 404, 395, 389, 370]

# Creating DataFrame
data_dict = {'Country': countries, 'Total Medals': total_medals}
df = pd.DataFrame(data_dict)

# Sorting the DataFrame by 'Total Medals' in ascending order for horizontal plotting
df = df.sort_values(by='Total Medals', ascending=True)

# Plotting
plt.figure(figsize=(8, 6))
bars = plt.barh(df['Country'], df['Total Medals'], color='#cca667')

# Adding the data labels next to the bars
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, 
             str(int(bar.get_width())), 
             va='center', ha='left', fontsize=6, fontweight='bold')

# Final plot adjustments
plt.xlabel('Total Medals')
plt.ylabel('Country')
plt.title('Total Olympic Medals by Country')

plt.tight_layout()
plt.show()
```

```python


# Data
countries = ['USA', 'China', 'Germany', 
             'Russian', 'France', 'Britain', 
             'Italy', 'Canada', 'Australia', 'Japan']
total_medals = [1344, 713, 667, 543, 436, 421, 404, 395, 389, 370]

# Creating DataFrame
data_dict = {'Country': countries, 'Total Medals': total_medals}
df = pd.DataFrame(data_dict)

# Sorting the DataFrame by 'Total Medals' in descending order to have the country with highest medals on top
df = df.sort_values(by='Total Medals', ascending=False)

# Plotting
plt.figure(figsize=(5, 6))
bars = plt.bar(df['Country'], df['Total Medals'], color='#cca667')

# Adding the data labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 30, yval, ha='center', va='bottom', fontsize=8)

# Final plot adjustments
plt.xlabel('Country')
plt.ylabel('Total Medals')
plt.title('Total Olympic Medals by Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

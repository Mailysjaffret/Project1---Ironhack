```python
# Average medal wins per country

#avg_medals = 
final_df.groupby('country_name')['combined_medals_count'].mean().sort_values(ascending=False)\
.astype(int).head(10)
```

```python
#checking for how this changes with sum

final_df.groupby('country_name')['combined_medals_count'].sum().sort_values(ascending=False)\
.astype(int).head(10)
```

```python
# looks like the top 10 performing countries are USA, Germany, Russian Federation, France, Italy
# Canada, Australia, Japan, Norway, Netherlands....the orders change if you do total sum vs avg
# but otherwise the same countries appear
```

```python
# this is a df of the top winners
top_winners_delta = final_df.groupby('country_name')['combined_medals_count'].sum().sort_values(ascending=False)\
.astype(int).head(10)
```

```python
# this is a list of the top winners

top_winners = top_winners_delta.index.tolist()
```

```python
# Average gdp & population per country - 'population' is rounded

final_df.groupby('country_name')[['gdp_pc', 'population']].mean().round().head()
```

```python
# making a full-length df out of the above

gdp_pop_df = final_df.groupby('country_name')[['gdp_pc', 'population']].mean().round()
```

```python
# making a fuller picture for the top winners, however medal count is not in this

top_winners_df = gdp_pop_df.loc[top_winners]
```

```python
top_winners_df
```

```python
# figuring out how to put together the final top winners picture df

#gdp_pop_df.loc[top_winners_delta]  -> tried this but then switched to merge
pd.merge(top_winners_delta, gdp_pop_df, on=['country_name'], how='left')
```

```python
# making a df with the above
top_winners_in_full_df = pd.merge(top_winners_delta, gdp_pop_df, on=['country_name'], how='left')
```
